```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

url = 'https://data.cdc.gov/api/views/2ew6-ywp6/rows.csv?accessType=DOWNLOAD'
csv_data = pd.read_csv(url)

wa_data = csv_data[ (csv_data['wwtp_jurisdiction'].str.contains("Washington")) | (wa_data['percentile'] == np.nan) ]
wa_data.head()
print(wa_data.count())
```

```python
king_county_data = wa_data[wa_data['county_names'].str.contains("King")]
kc_data_sorted=king_county_data.sort_values(by='date_end') 
kc_data_sorted.head()
kc_data_sorted.tail()
kc_data_sorted.county_names.unique()
kc_data_sorted.index = kc_data_sorted['date_end']
```

```python
#king_county_data.date_end = pd.to_datetime(king_county_data.date_end)
kc_data_sorted.index = kc_data_sorted['date_end']
#wa_data.index = wa_data['date_end']
```

```python
kp_data = kc_data_sorted[kc_data_sorted['county_names'].str.contains("King,Pierce")]
kp_data = kp_data.set_index('date_end')

ks_data = kc_data_sorted[kc_data_sorted['county_names'].str.contains("King,Snohomish")]
ks_data = ks_data.set_index('date_end')
```

```python
kp_data = kc_data_sorted[kc_data_sorted['county_names'].str.contains("King,Pierce")]
kp_data = kp_data[['date_end', 'percentile']]
kp_data['date_end'] = pd.to_datetime(kp_data['date_end'])
kp_data = kp_data.set_index('date_end')

ks_data = kc_data_sorted[kc_data_sorted['county_names'].str.contains("King,Snohomish")]
ks_data = ks_data[['date_end', 'percentile']]
ks_data['date_end'] = pd.to_datetime(ks_data['date_end'])
ks_data = ks_data.set_index('date_end')
ks_data_mean = ks_data.groupby('date_end').mean()

```

```python

fig, ax = plt.subplots()

plt.plot(kp_data.index, kp_data['percentile'])
plt.plot(ks_data_mean.index, ks_data_mean['percentile'])

plt.xlabel("Date")
plt.xticks(rotation=45, ha='right')
plt.ylabel("King County Wastewater %ile")
plt.yticks(np.arange(0, 100, step=5))
plt.ylim(0, 100)
plt.legend(['King,Pierce', 'King,Snohomish'])
plt.show()
```

```python
!pip install plotly
import plotly.graph_objs as go

# create traces for each county
trace1 = go.Scatter(x=kp_data.index, y=kp_data['percentile'], name='King,Pierce')
trace2 = go.Scatter(x=ks_data_mean.index, y=ks_data_mean['percentile'], name='King,Snohomish')

# create the figure with a date slider
fig = go.Figure(data=[trace1, trace2],
        layout=go.Layout(
          xaxis=dict(
            rangeselector=dict(
              buttons=list([
                dict(count=1,
                   label='1m',
                   step='month',
                   stepmode='backward'),
                dict(count=6,
                   label='6m',
                   step='month',
                   stepmode='backward'),
                dict(count=1,
                   label='YTD',
                   step='year',
                   stepmode='todate'),
                dict(count=1,
                   label='1y',
                   step='year',
                   stepmode='backward'),
                dict(step='all')
              ])
            ),
            rangeslider=dict(
              visible=True
            ),
            type='date'
          )
        ))

fig.update_layout(title='King County Wastewater %ile',
          xaxis_title='Date',
          yaxis_title='King County Wastewater %ile')

fig.show()
```

```python
import ipywidgets as widgets
from IPython.display import display

# get the unique county names from the csv_data dataframe
county_names = csv_data['county_names'].unique()

# create a dropdown widget with the county names
county_dropdown = widgets.Dropdown(
    options=county_names,
    value=county_names[0],
    description='County:',
    layout=widgets.Layout(width='50%')
)

# define a function to update the plot based on the selected county
def update_plot(county):
    # get the data for the selected county
    data = csv_data[csv_data['county_names'] == county][['date_end', 'percentile']]
    data['date_end'] = pd.to_datetime(data['date_end'])
    data = data.set_index('date_end')
    
    # check if the trace for the selected county already exists in the plot
    trace_exists = False
    for trace in fig.data:
        if trace.name == county:
            trace_exists = True
            trace.x = data.index
            trace.y = data['percentile']
    
    # if the trace doesn't exist, create a new one
    if not trace_exists:
        new_trace = go.Scatter(x=data.index, y=data['percentile'], name=county)
        fig.add_trace(new_trace)
    
    # update the layout of the plot
    fig.update_layout(title=f'{county} Wastewater %ile',
                      xaxis_title='Date',
                      yaxis_title=f'{county} Wastewater %ile')
    
    # display the updated plot
    fig.show()

# create an initial trace for the first county in the county_names list
data = csv_data[csv_data['county_names'] == county_names[0]][['date_end', 'percentile']]
data['date_end'] = pd.to_datetime(data['date_end'])
data = data.set_index('date_end')
trace = go.Scatter(x=data.index, y=data['percentile'], name=county_names[0])

# create the figure with the initial trace
fig = go.Figure(trace)

# link the dropdown widget to the update function
widgets.interactive(update_plot, county=county_dropdown)

# display the dropdown widget
display(county_dropdown)
```
